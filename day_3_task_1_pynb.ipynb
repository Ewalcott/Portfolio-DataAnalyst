```python
#q1 Write the code to read a CSV file into a Pandas DataFrame.
import pandas as pd

# Read a local csv file
df = pd.read_csv('student.csv')
```

exercise 2 Indexing and Sclicing

```python
#q1 Write the code to select the 'name' column

# this code Selects the column as a Pandas
name_column = df['name']
print(name_column)
```

```python
#q2 Write the code to select the 'name' and 'mark' columns

# Selects all rows for 'name' and 'mark' columns
df_subset = df.loc[:, ['name', 'mark']]
print(df_subset)
```

```python
#q3 Write the code to select the first 3 rows

# Selects rows at index data 0, 1, and 2
first_three_rows = df.iloc[0:3]
print(first_three_rows)
```

```python
#q4 Write the code to select all rows where the 'class' is 'Four

# selects rows where 'class' = 'Four'
class_four_df = df.loc[df['class'] == 'Four']
print(class_four_df)
```

Excercise 3 Data Manupliation

```python
import numpy as np

#q1 Write the code to add a new column 'passed' that indicates whether the student passed (mark >= 60).

# If mark is 60 or higher, 'passed'= True, if not False
df['passed'] = df['mark'] >= 60
print(df)
```

```python
#q2 Write the code to rename the 'mark' column to 'score

# to Rename 'mark' to 'score'
df = df.rename(columns={'mark': 'score'})
print(df)
```

```python
#q3 Write the code to drop the 'passed' column

# this code will
#Drop meaning remove the 'passed' column
df = df.drop(columns=['passed'])
print(df)
```

Exercise 4: Aggregation and Grouping

```python
#q1 Write the code to group the DataFrame by the 'class' column and calculate the mean 'score’ for each group.

# 1 is to Group by 'class'
# 2 is to Select the 'score' column
# 3 to Calculate the mean of group
class_mean = df.groupby('class')['score'].mean()

print(class_mean)
```

```python
#q2 Write the code to count the number of students in each class.

# to Group by 'class'
#and count how many students are in each group
student_counts = df.groupby('class')['name'].count()

print(student_counts)
```

```python
#q3 Write the code to calculate the average score for each gender.

# Group by 'gender',
#calcullate 'score', by average
gender_averages = df.groupby('gender')['score'].mean()

print(gender_averages)
```

Exercise 5: Exporting Data

```python
#q1 Write the code to save the DataFrame with the new 'grade' column to a new CSV file.

# Save the DataFrame to a new file named 'student_grades.csv'
df.to_csv('student_grades.csv', index=False)
```

```python
#q7 Visualising data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import files
import os # To check if file exists

# Load the original CSV (now guaranteed to be present)
df = pd.read_csv('student.csv')

# Apply transformations as per previous cells to reconstruct the DataFrame state
df['passed'] = df['mark'] >= 60
df = df.rename(columns={'mark': 'score'})
df = df.drop(columns=['passed'])

# 2. Preview the first 5 rows to ensure it loaded correctly
print(df.head())

# 3. Create a Visualization (e.g., a scatter plot)
# Using 'id' for x-axis, 'score' for y-axis, and 'gender' for hue.
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='id', y='score', hue='gender')

# 4. Add titles and labels
plt.title('Student Scores by ID and Gender')
plt.xlabel('Student ID')
plt.ylabel('Score')
plt.legend(title='Gender')
plt.grid(True)
plt.show()
```
